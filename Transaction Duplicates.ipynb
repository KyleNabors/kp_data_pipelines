```python
import pandas as pd

df = pd.read_csv("C:\\Users\\O304312\\Downloads\\Transaction.csv")

df = df.drop(columns=["Unnamed: 0"])
```

```python
# select all rows that have exact duplicates (including the first occurrence)
dupes = df[df.duplicated(keep=False)].copy()

# create a stable key for the full row (so truly identical rows sort together)
key_cols = list(df.columns)
dupes["_dup_key"] = dupes[key_cols].astype(str).agg("|".join, axis=1)

# sort by that key (and then by columns if you want) so identical rows appear next to each other
dupes = dupes.sort_values(by=["_dup_key"] + key_cols).reset_index(drop=True)

# optional: add a column that shows how many copies of each exact row exist
dupes["duplicate_count"] = dupes.groupby("_dup_key")["_dup_key"].transform("count")

# export for review
dupes.to_excel(
    "C:\\Users\\O304312\\Downloads\\Transaction Duplicates.xlsx", index=False
)
```
