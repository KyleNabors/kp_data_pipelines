```python
trac_raw = pd.read_json(r"J:\TRAC\TRAC_Data.json", encoding="utf-16")

# expand the nested dicts into a new DataFrame
trac = pd.json_normalize(trac_raw["TRAC_Data"])

# keep the original index if needed
trac.index = trac_raw.index

trac = trac[["StudyTitle", "FullyExecutedDate"]]


# inspect
trac.info()
trac.head()
```

```python
vp_studies = vp_studies.merge(
    trac, left_on="Site Study Code", right_on="StudyTitle", how="left"
)
vp_studies = vp_studies.drop_duplicates()
```
