```python
import pandas as pd
```

```python
# Appian Data

appian = pd.read_excel(
    r"C:\Users\O304312\Downloads\Appian Study Startup SQL.xlsx",
)
```

```python
legacy_Study = pd.read_excel(r"C:\Users\o304312\Downloads\Legacy Task Data.xlsx")
legacy_Study["TaskStartDate"] = pd.to_datetime(legacy_Study["TaskStartDate"])
legacy_Study = legacy_Study[
    legacy_Study["TaskStartDate"] < pd.to_datetime("2024-11-30")
]

legacy_Study["isLegacy"] = 1
legacy_Study["RowID"] = range(len(appian) + 1, len(appian) + len(legacy_Study) + 1)
```

```python
legacy_Study.rename(
    columns={
        "LegacyStudyID": "StudyID",
        "StudyPrincipalInvestigator": "Study Principal Investigator",
        "RCPProjectTitle": "RCP Project Title",
        "DaysFromRCAToContractExecution": "Days Since RCA Task",
        "StudyStatus": "Study Status",
        "TaskOwner": "Task Owner",
        "AssignedGroup": "Task Assigned Group",
        "TaskStartDate": "Task Start Date",
        "TaskCompletedDate": "Task Completed Date",
    },
    inplace=True,
)
```

```python
appian["KPOCTStudy"] = 0
appian["isLegacy"] = 0
appian["RowID"] = range(1, len(appian) + 1)
appian["Days Since RCA Task"] = None
```

```python
merged = pd.concat([appian, legacy_Study], join="inner", ignore_index=True)

merged = pd.merge(
    merged,
    appian[["RowID", "Study Closed Date"]],
    on="RowID",
    how="left",
    indicator=True,
)
```

```python
merged.to_excel(
    r"C:\Users\O304312\Downloads\Merged Appian and Legacy Data.xlsx",
)
```
